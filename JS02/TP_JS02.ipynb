```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

print("=== Langkah 1: Load dataset & hapus kolom yang tidak digunakan ===")

# Baca dataset
df = pd.read_csv("wbc.csv")

# Lihat 5 baris pertama
print("\nData awal:")
print(df.head())

# Hapus kolom 'id' (tidak digunakan)
if 'id' in df.columns:
    df = df.drop(columns=['id'])
```

```python
from sklearn.preprocessing import LabelEncoder

# Buat objek LabelEncoder
le = LabelEncoder()

# Transform kolom diagnosis
df['diagnosis'] = le.fit_transform(df['diagnosis'])

print("Setelah encoding diagnosis:")
print(df[['diagnosis']].head(10))
print("\nMapping LabelEncoder:", dict(zip(le.classes_, le.transform(le.classes_))))
```

```python
from sklearn.preprocessing import StandardScaler

print("=== Langkah 4: Standardisasi semua kolom numerik ===")

# Buat scaler
scaler = StandardScaler()

# Pilih semua kolom numerik kecuali diagnosis
numerical_cols = df.drop(columns=['diagnosis']).select_dtypes(include=['float64','int64']).columns

# Standardisasi kolom numerik
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("\n5 baris pertama setelah standardisasi:")
print(df.head())

# Cek statistik mean & std
print("\nStatistik setelah standardisasi (mean & std):")
print(df[numerical_cols].describe().T[['mean','std']].head(10))
```
