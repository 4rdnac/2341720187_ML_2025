# Tugas
## Load Dataset

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Muat dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Clustering dengan HDBSCAN

```python
import hdbscan

# Lakukan clustering
clusterer = hdbscan.HDBSCAN(min_cluster_size=5)
cluster_labels = clusterer.fit_predict(X_scaled)
```

## Jumlah Cluster dan Noise yang Terbentuk

```python
import numpy as np

n_clusters = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
n_noise = np.sum(cluster_labels == -1)

print(f"Jumlah cluster yang terbentuk: {n_clusters}")
print(f"Banyaknya noise: {n_noise}")
```

## Visualisasi

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduksi dimensi menjadi 2 komponen utama
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot hasil clustering
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='rainbow', s=50)
plt.title("Visualisasi Hasil Clustering HDBSCAN pada Dataset Iris (PCA 2D)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

## Kesimpulan
Hasil clustering menghasilkan 2 cluster utama dan 2 data dianggap noise. Hal ini menunjukkan bahwa HDBSCAN mendeteksi dua kelompok besar yang cukup jelas terpisah.
